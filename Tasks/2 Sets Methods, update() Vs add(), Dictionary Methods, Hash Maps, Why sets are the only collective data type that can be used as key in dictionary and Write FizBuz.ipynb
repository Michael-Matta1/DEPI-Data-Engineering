# Sets Methods


add()	Adds a given element to a set

clear()	Removes all elements from the set

copy()	Returns a shallow copy of the set

difference()	Returns a set that is the difference between two sets

difference_update()	Updates the existing caller set with the difference between two sets

discard()	Removes the element from the set

frozenset()	Return an immutable frozenset object

intersection()	Returns a set that has the intersection of all sets

intersection_update()	Updates the existing caller set with the intersection of sets

isdisjoint()	Checks whether the sets are disjoint or not

issubset()	Returns True if all elements of a set A are present in another set B

issuperset()	Returns True if all elements of a set A occupies set B

pop()	Returns and removes a random element from the set

remove()	Removes the element from the set

symmetric_difference()	Returns a set which is the symmetric difference between the two sets

symmetric_difference_update()	Updates the existing caller set with the symmetric difference of sets

union()	Returns a set that has the union of all sets

update()	Adds elements to the set

````python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

set1.add(5)
print("add():", set1)

set1.clear()
print("clear():", set1)

set1 = {1, 2, 3, 4}
set1_copy = set1.copy()
print("copy():", set1_copy)

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

diff = set1.difference(set2)
print("difference():", diff)

set1.difference_update(set2)
print("difference_update():", set1)

set1.discard(2)
print("discard():", set1)

set1 = {1, 2, 3, 4}
intersection = set1.intersection(set2)
print("intersection():", intersection)

set1 = {1, 2, 3, 4}
set1.intersection_update(set2)
print("intersection_update():", set1)

print("isdisjoint():", set1.isdisjoint(set2))
print("issubset():", set1.issubset(set2))

print("issuperset():", set2.issuperset(set1))

elem = set1.pop()
print("pop():", elem, set1)

set1 = {1, 2, 3, 4}

set1.remove(2)
print("remove():", set1)

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
sym_diff = set1.symmetric_difference(set2)
print("symmetric_difference():", sym_diff)

set1.symmetric_difference_update(set2)
print("symmetric_difference_update():", set1)

union_set = set1.union(set2)
print("union():", union_set)

set1.update({7, 8})
print("update():", set1)
````

### Hash Tables and Dictionaries
Python dictionaries are implemented using a data structure called a hash table. Here’s a basic rundown of how hash tables work:
1. **Hash Function**: A hash function takes an input (the dictionary key) and returns a number (the hash value). This number is used to determine where the key-value pair is stored in the hash table.
2. **Storage**: The hash value determines the "bin" or "bucket" in the hash table where the key-value pair will be stored.

### Why Keys Must Be Immutable
The immutability requirement for dictionary keys stems from how hash tables work. Here’s why:

1. **Consistency of Hash Value**: When a key is first added to a dictionary, its hash value is calculated and used to place it in the appropriate bin. If the key were mutable and its content (value) changed, its hash value would also change. This creates a problem because:
   - **Lookup Issue**: The dictionary would no longer be able to find the key in the correct bin, as the hash value used to place it there has changed. Therefore, even though the key still exists in the dictionary, you wouldn't be able to find it using the new value.

2. **Key Integrity**: Because the key’s hash value determines its position in the hash table, allowing the key to change while it is in use can break the dictionary’s structure:
   - **Inconsistency**: The dictionary might have a key-value pair stored in a location determined by an old hash value, while the new hash value points to a different location. This can lead to inconsistencies and errors in data retrieval.

### Illustration
Imagine a simple scenario where you use a list as a dictionary key:
1. **Initial State**:
   ```python
   my_dict = {}
   key = [1, 2, 3]
   my_dict[key] = "value"
   ```
   - The hash value of `[1, 2, 3]` is calculated.
   - The key-value pair (`[1, 2, 3]`: "value") is stored in the hash table based on this hash value.

2. **Mutation**:
   ```python
   key.append(4)  # Now key is [1, 2, 3, 4]
   ```
   - The contents of `key` have changed.
   - The hash value of `[1, 2, 3, 4]` is different from the hash value of `[1, 2, 3]`.

3. **Lookup Failure**:
   ```python
   my_dict[key]  # This will fail to find the key
   ```
   - The dictionary will use the new hash value of `[1, 2, 3, 4]` to look up the key, but the key was stored under the old hash value of `[1, 2, 3]`.

To prevent these issues, dictionary keys are required to be immutable. Immutable objects (like strings, numbers, and tuples) have a fixed hash value because their content cannot change. This ensures the integrity and consistency of the dictionary’s structure.

In the context of hash tables (and thus Python dictionaries), "bin" or "bucket" refers to a specific location in the hash table where key-value pairs are stored. Here's a more detailed explanation:

### Hash Tables and Bins/Buckets
A hash table is an array-like data structure that stores key-value pairs. When you add a key-value pair to a dictionary, the hash table uses a hash function to compute a hash value from the key. This hash value is then used to determine the index (or address) in the array where the key-value pair should be stored. This index is called a "bin" or "bucket".

#### Why Bins/Buckets?
- **Efficiency**: Using hash values to determine the storage location allows for very fast data retrieval. Once you have the hash value, you can directly access the bin where the data is stored.
- **Collisions**: Sometimes, different keys can produce the same hash value. This is called a collision. Hash tables handle collisions by storing multiple key-value pairs in the same bin. There are various strategies to handle collisions, such as chaining (where each bin is a linked list of key-value pairs) or open addressing (where the algorithm finds another open bin based on certain rules).

### Visual Representation
Let's illustrate with a simple example:

1. **Hash Function**: A hash function maps keys to hash values.
   - Example: `hash("apple") = 42`
   - Example: `hash("banana") = 73`

2. **Bins/Buckets**: Imagine the hash table as an array with 100 bins (indexed from 0 to 99).

   ```
   Hash Table (simplified visualization):
   Index  | Bin Content
   -------|------------
   0      | ...
   1      | ...
   2      | ...
   ...
   42     | ("apple", "fruit")
   ...
   73     | ("banana", "fruit")
   ...
   99     | ...
   ```

3. **Storing Data**:
   - The key `"apple"` is hashed to the value `42`, so the key-value pair `("apple", "fruit")` is stored in bin 42.
   - The key `"banana"` is hashed to the value `73`, so the key-value pair `("banana", "fruit")` is stored in bin 73.

4. **Retrieving Data**:
   - To retrieve the value for the key `"apple"`, the hash table calculates the hash value `42` and looks in bin 42.

By organizing data into bins/buckets based on hash values, hash tables (and dictionaries) can efficiently store and retrieve key-value pairs. This mechanism underlies the requirement that dictionary keys must be immutable: if keys could change, their hash values (and thus their storage bins) would also change, leading to inconsistencies and lookup failures.

   - When you try to look up the same object (using `key1`) in the dictionary, Python calculates a new hash value based on the modified key. However, this new hash value points to a different location in the hash table.
   - As a result, Python won't find the entry associated with the original `key1` because it's now in a different hash bin.

4. **Old Value Lookup**:
   - Suppose you had stored a value associated with the original `key1` (let's call it `value1`).
   - If you tried to look up the old value (using `key1`), it wouldn't be found either. Why? Because the hash value now points to a different location, and the value stored there might be something else (not `value1`).

In summary, using immutable keys ensures consistency in dictionary lookups. Mutable keys could lead to unexpected behavior due to changing hash values. That's why Python enforces immutability for dictionary keys. 🐍
